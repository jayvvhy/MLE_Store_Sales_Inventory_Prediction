#### EDA on holidays

```python
holidays = pd.read_csv(r"K:\SMU August 2025\CS611-G1-Machine Learning Engineering\Group Project -5\data\store-sales-time-series-forecasting\holidays_events.csv")
holidays.head(10)
#holidays.isnull().sum()
```

##### Holiday file has no NaN columns(No null values.)

```python
holidays.info()
holidays.duplicated().sum()
```

```python
col_to_plot = ['type', 'locale']


for col in col_to_plot:
    # Count the occurrences of each category
    counts = holidays[col].value_counts()

    plt.figure(figsize=(18, 8))
    plt.bar(counts.index, counts.values, color='skyblue', edgecolor='black')
    plt.title(f'Frequency Distribution of {col} column')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xticks(rotation=90)
    plt.tight_layout()  # adjust layout so labels fit nicely
    plt.show()
```
