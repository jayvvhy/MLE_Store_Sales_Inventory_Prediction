```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.multioutput import RegressorChain
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
```

```python
store = pd.read_csv(r"K:\SMU August 2025\CS611-G1-Machine Learning Engineering\Group Project -5\data\store-sales-time-series-forecasting\stores.csv")
store.head()
```

```python
store.info()
store.duplicated().sum()
store.isnull().sum()
```

```python
# For categorical columns only
col_to_plot = ['city', 'state', 'type']

for col in col_to_plot:
    # Count the occurrences of each category
    counts = store[col].value_counts()

    plt.figure(figsize=(18, 8))
    plt.bar(counts.index, counts.values, color='skyblue', edgecolor='black')
    plt.title(f'Frequency Distribution of {col} column')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.tight_layout()  # adjust layout so labels fit nicely
    plt.show()
```
