```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('data/train.csv')
store_df = pd.read_csv('data/stores.csv')
oil_df = pd.read_csv('data/oil.csv')
holiday_df = pd.read_csv('data/holidays_events.csv')
transaction_df = pd.read_csv('data/transactions.csv')
```

```python
dfs = {
    'train_df': train_df,
    'store_df': store_df,
    'oil_df': oil_df,
    'holiday_df': holiday_df,
    'transaction_df': transaction_df
}

# Loop through and print date ranges
for name, df in dfs.items():
    if 'date' in df.columns:
        df['date'] = pd.to_datetime(df['date'])
        min_date = df['date'].min()
        max_date = df['date'].max()
        print(f"{name}: {min_date.date()} → {max_date.date()}")
    else:
        print(f"{name}: ❌ no 'date' column found")
```

```python
df = train_df.copy()
df = df.sort_values(['store_nbr', 'family', 'date'])

# Grouped by store & family
df['sales_lag_1'] = df.groupby(['store_nbr', 'family'])['sales'].shift(1)
df['sales_lag_7'] = df.groupby(['store_nbr', 'family'])['sales'].shift(7)
df['rolling_mean_7'] = df.groupby(['store_nbr', 'family'])['sales'].shift(1).rolling(7).mean()
df['rolling_std_7'] = df.groupby(['store_nbr', 'family'])['sales'].shift(1).rolling(7).std()

# Date features
df['date'] = pd.to_datetime(df['date'])
df['day_of_week'] = df['date'].dt.dayofweek
df['month'] = df['date'].dt.month

# Targets for next 7 days (for Option B)
for i in range(1, 8):
    df[f'y_d{i}'] = df.groupby(['store_nbr', 'family'])['sales'].shift(-i)
```

```python
import pandas as pd

# Create a sample DataFrame
data = {'A': [1, 2, 3, 4], 'B': [10, 20, 30, 40]}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Shift column 'A' down by one period
shifted_a = df['A'].shift(periods=1)
print("\nShifted column 'A':")
print(shifted_a)

# Shift the entire DataFrame up by one period
shifted_df_up = df.shift(periods=-1)
print("\nDataFrame shifted up:")
print(shifted_df_up)

# Shift the entire DataFrame to the right by one period
shifted_df_right = df.shift(periods=1, axis=1, fill_value=0)
print("\nDataFrame shifted right:")
print(shifted_df_right)
```
